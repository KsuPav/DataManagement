# 3. Построить запрос: Владелец Аэропорта, Аэропорт, Пассажиропоток суммарный за 2018, грузопоток суммарный за 2018.

```python
passenger_transportation = passenger_transportation.fillna(0)
passenger_transportation = passenger_transportation.replace('***', 0)

# Список месяцев
months = list(passenger_transportation.columns)
months.pop(0)
months.pop(0)

# Преобразование значений пассажиропотока в вещественный тип
for mon in months:
  passenger_transportation[mon] = passenger_transportation[mon].apply(lambda x: str(x).replace(' ', ''))
  passenger_transportation[mon] = passenger_transportation[mon].astype(float)

# Вычисление годовых значений пассажиропотока
passenger_transportation['Январь - Декабрь'] = passenger_transportation.iloc[:, 2:-1].sum(axis=1)

passenger_transportation
```

```python
cargo_transportation = cargo_transportation.fillna(0)
cargo_transportation = cargo_transportation.replace('***', 0)

# Список месяцев
months = list(cargo_transportation.columns)
months.pop(0)
months.pop(0)

# Преобразование значений грузопотока в вещественный тип
for mon in months:
  cargo_transportation[mon] = cargo_transportation[mon].apply(lambda x: str(x).replace(' ', ''))
  cargo_transportation[mon] = cargo_transportation[mon].apply(lambda x: str(x).replace(',', '.'))
  cargo_transportation[mon] = cargo_transportation[mon].astype(float)

# Вычисление годовых значений грузопотока
cargo_transportation['Январь - Декабрь'] = cargo_transportation.iloc[:, 2:-1].sum(axis=1)

cargo_transportation
```

```python
# Список номеров строк, в которых наименование аэропорта мужно срезать до пробела
by_space = [1, 2, 3, 4, 5, 6, 8, 9, 10, 13, 14, 15, 16, 17]

for i in by_space:
  ind = airports.loc[i, 'Наименование аэропорта РФ'].index(' ')
  airports.loc[i, 'Наименование аэропорта РФ'] = airports.loc[i, 'Наименование аэропорта РФ'][:ind]

# Ручная корректировка некоторых наименований аэропортов
airports.loc[7, 'Наименование аэропорта РФ'] = 'Нижняя Пеша'
airports.loc[11, 'Наименование аэропорта РФ'] = 'Летняя  Золотица'
airports.loc[12, 'Наименование аэропорта РФ'] = 'Верхняя Золотица'
airports.loc[18, 'Наименование аэропорта РФ'] = 'Каменка'
airports.loc[19, 'Наименование аэропорта РФ'] = 'Несь'
airports.loc[210, 'Наименование аэропорта РФ'] = 'Ярославль (Туношна)'
airports.loc[228, 'Наименование аэропорта РФ'] = 'Челябинск'

# Удаление ненужных пробелов из наименований аэропортов
airports['Наименование аэропорта РФ'] = airports['Наименование аэропорта РФ'].apply(lambda x: str(x).replace(' (', '('))
airports['Наименование аэропорта РФ'] = airports['Наименование аэропорта РФ'].apply(lambda x: str(x).rstrip().lstrip())
```

```python
airports
```

```python
# Создание датафрейма под запрос
request = pd.DataFrame()

# Добавление стобцов 'Юрлицо' и 'Наименование аэропорта РФ' из airports
request['Юрлицо'] = airports['Юрлицо']
request['Наименование аэропорта РФ'] = airports['Наименование аэропорта РФ']

# Извлечение данных о пассажиропотоке за 2018 год
pt_2018 = passenger_transportation.loc[passenger_transportation['Год периода данных'] == 2018]

# Наполнение request соответствующими данными о пассажиропотоке за 2018 год
# try/except пропускает аэропорты, о пассажиропотоке которых нет информации
for i in list(request['Наименование аэропорта РФ']):
  try:
    request.loc[(request['Наименование аэропорта РФ'] == i), 'Пассажиропоток суммарный за 2018'] = int(pt_2018.loc[pt_2018['Наименование аэропорта РФ'] == i]['Январь - Декабрь'])
  except:
    continue


# Извлечение данных о грузопотоке за 2018 год
ct_2018 = cargo_transportation.loc[cargo_transportation['Год периода данных'] == 2018]

# Наполнение request соответствующими данными о грузопотоке за 2018 год
# try/except пропускает аэропорты, о грузопотоке которых нет информации
for i in list(request['Наименование аэропорта РФ']):
  try:
    request.loc[(request['Наименование аэропорта РФ'] == i), 'Грузопоток суммарный за 2018'] = int(ct_2018.loc[ct_2018['Наименование аэропорта РФ'] == i]['Январь - Декабрь'])
  except:
    continue
```

# 4. Перечислить аэропорты, где пассажиропоток больше медианы, а грузопоток меньше медианы.

```python
# В new собираются ненулевые данные за 2018 год
new = request[request['Пассажиропоток суммарный за 2018'].notnull()]
new = new.loc[(new['Пассажиропоток суммарный за 2018'] != 0) & (new['Грузопоток суммарный за 2018'] != 0)]

new
```

```python
# Медианы пассажиропотка и грузопотока среди суммарных значений аэропортов за 2018 год
pt_median = new['Пассажиропоток суммарный за 2018'].median()
ct_median = new['Грузопоток суммарный за 2018'].median()

# Аэропорты, где пассажиропоток больше медианы, а грузопоток меньше медианы
result = new.loc[(new['Пассажиропоток суммарный за 2018'] > pt_median) & (new['Грузопоток суммарный за 2018'] < ct_median)]

print('Аэропорты, где пассажиропоток больше медианы, а грузопоток меньше медианы:\n', list(result['Наименование аэропорта РФ']))
```

# 5. Перечислить авиакомпании у которых нет типов воздушных судов зарегистрированных не в России

```python
airlines = airlines[airlines['Самолеты'].notna()]
airlines
```

```python
# Список уникальных типов самолетов зарегистрированных в России
print(unique_rus_aircrafts)
```

```python
airl = []

for index, row in airlines.iterrows():
  answer = True # True - все самолеты текущей авиакомпании зарегистрированы в России, иначе False
  planes = row['Самолеты'].split(',')
  
  for i in planes:
    # Убираем из названий самолетов лишнее
    i = i.lstrip()
    ind = i.index(' ')
    i = i[:ind]
    # Если текущего самолета нет в списке уникальных самолетов, зарегистрированных в России, то answer становится False
    answer = answer and (i in unique_rus_aircrafts)
  
  # Сохраняем авиакомпанию, если все ее самолеты зарегистрированы в России
  if answer:
    airl.append(airlines.loc[index]['Индекс'])

print('Авиакомпании у которых нет типов воздушных судов зарегистрированных не в России:\n', airl)
```

# 6. Выведите список: Месяц, суммарный пассажиропоток за данный месяц, аэропорт, в котором пассажиропоток в данном месяце минимальный

```python
# В min_pt собираются ненулевые данные за 2007 год, т.к. по этому году наиболее полная информация
min_pt = passenger_transportation.loc[(passenger_transportation['Год периода данных'] == 2007) & (passenger_transportation['Январь - Декабрь'] != 0)]
# Список месяцев
months.pop(-1)
```

```python
# Для каждого месяца определяется минимальный пассажиропоток и аэропорт с этим минимальным пассажиропотоком
for i in months:
  mini = min_pt[i].min()
  x = list(min_pt.loc[min_pt[i] == mini]['Наименование аэропорта РФ'])[0]
  print(f'Месяц: {i} \nСуммарный пассажиропоток за данный месяц: {round(min_pt[i].sum(), 2)} \nАэропорт с минимальным пассажиропотоком: {x} (пассажиропоток: {mini})\n')
```

# 7. Вывести список: тип аэропорта, средний грузопоток в месяц в аэропортах данного типа

```python
request['Тип'] = airports['Тип'].apply(lambda x: str(x).replace('  ', ' ').rstrip().lstrip())
request = request[request['Грузопоток суммарный за 2018'].notna()]

request
```

```python
# Группировка по типу и вычисление среднего внутри групп
averages = request.groupby('Тип')['Грузопоток суммарный за 2018'].mean()/12
print(averages)
```
